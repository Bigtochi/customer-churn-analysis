This project focused on analyzing customer behavior data to understand the factors that may contribute to customer churn. Using exploratory data analysis and visualizations, I examined customer patterns and behaviors to identify trends associated with customers leaving a service.
From the analysis, I found that customers who churned generally had a higher number of support calls compared to retained customers. I also observed that payment delays appeared to have a relationship with churn, while customer usage frequency and tenure showed noticeable differences between customers who stayed and those who left.
Based on these findings, businesses could improve customer retention by strengthening customer support processes, reducing payment-related issues, and identifying customers with lower activity levels before they decide to leave. Early intervention strategies can also help reduce churn among high-risk customers.
Overall, this analysis highlights how customer behavior can provide valuable insights into retention patterns and demonstrates how data can be used to support better business decisions.
